```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.colors as mcolors

def compare_decimal_data(data1_str, data2_str):
    # convert strings to lists of integers
    data1 = list(map(int, data1_str.strip().split()))
    data2 = list(map(int, data2_str.strip().split()))

    # Ensure both data lists are of the same length
    if len(data1) != len(data2):
        print(f"Data lengths are different! Data1: {len(data1)}, Data2: {len(data2)}")
        return
    
    error_map = np.zeros(len(data1))  # Array to hold errors
    
    # Loop through to find errors
    differences = 0
    for i in range(len(data1)):
        if data1[i] != data2[i]:
            error_map[i] = 1  # Mark position with an error
            differences += 1

    print(f"Total number of differences: {differences}")
    print(f"Bit error rate: {differences / len(data1) * 100:.2f}%")
    print(error_map)

    return error_map  # Return the array of errors

def plot_error_heatmap(error_map, test_name="Test"):
    plt.figure(figsize=(20, 2))
    plt.imshow(error_map[np.newaxis, :], cmap='summer', aspect='auto') 
    plt.colorbar(label="Error Presence (1 = Error)")
    plt.title(f"Bit Error Heatmap for {test_name}")
    plt.yticks([]) 
    plt.xlabel("Bit Position")
    plt.show()

def plot_error_heatmap_with_packets(error_map, packet_size, test_name="Test"):
    error_map_np = np.array(error_map)
    plt.figure(figsize=(40, 4)) 

    cmap = mcolors.ListedColormap(['cornflowerblue', 'gold'])

    plt.imshow(error_map_np[np.newaxis, :], cmap=cmap, aspect='auto',vmin=0, vmax=1, interpolation='none') 
    cbar = plt.colorbar(label="0 = No Error, 1 = Error", location='left', pad=0.02, ticks=[0, 1]) 
    cbar.ax.tick_params(labelsize=14) 
    cbar.set_label("0 = No Error, 1 = Error", fontsize=14)
    plt.title("Bit Errors", fontsize=22)

    # Calculate packet boundaries
    num_packets = len(error_map) // packet_size
    remainder = len(error_map) % packet_size

    # Draw lines between full packets
    for i in range(1, num_packets):
        plt.axvline(x=i * packet_size - 0.5, color='black', linestyle='--')

    if remainder > 0:
        plt.axvline(x=num_packets * packet_size - 0.5, color='black', linestyle='--')

    # Add packet numbers as x-axis tick labels
    num_packets = len(error_map) // packet_size + (1 if len(error_map) % packet_size != 0 else 0)
    packet_indices = [i * packet_size for i in range(num_packets)] 
    packet_labels = [f"{i}" for i in range(num_packets)] 
    
    plt.xticks(packet_indices, packet_labels, fontsize=14)  # Set the packet numbers as x-ticks


    plt.yticks([])  # Remove y-axis ticks
    plt.xlabel("Packet Number", fontsize=20)
    plt.show()



def average_errors_over_tests(data_pairs):
    total_errors = None
    num_tests = len(data_pairs)

    for idx, (data1_str, data2_str) in enumerate(data_pairs):
        error_map = compare_decimal_data(data1_str, data2_str)
        
        if total_errors is None:
            total_errors = error_map
        else:
            total_errors += error_map  # Accumulate errors

    avg_errors = total_errors / num_tests  # Calculate average errors
    return avg_errors


```

```python
import re
import numpy as np

# Define packet sizes based on SF values
sf_packet_sizes = {
    7: 250,
    8: 250,
    10: 50,
    12: 50
}

def parse_experiments(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

    # extracting experiment data
    experiment_pattern = r'Experiment with SF: (\d+), CR: (\d+), Tx Power: (\d+), Toa: (\d+):\s*([\d\s]+)'
    experiments = re.findall(experiment_pattern, content)

    parsed_data = []
    
    for experiment in experiments:
        sf, cr, tx_power, toa, data_block = experiment
        
        # Split into list of integers
        data = list(map(int, data_block.split()))
        
        # Store parsed data
        parsed_data.append({
            'SF': int(sf),
            'CR': int(cr),
            'Tx Power': int(tx_power),
            'ToA': int(toa),
            'data': data
        })
    
    return parsed_data


def compare_and_plot_experiments_with_dynamic_packet_size(file_path, reference_data_str):
    reference_data = list(map(int, reference_data_str.split()))
    experiments = parse_experiments(file_path)

    # Loop through each experiment, compare the data, generate heatmap
    for experiment in experiments:
        sf = experiment['SF']
        cr = experiment['CR']
        tx_power = experiment['Tx Power']
        toa = experiment['ToA']
        data = experiment['data']

        packet_size = sf_packet_sizes.get(sf, 100)  # Default to 100 if SF is not in the dictionary

        print(f"Experiment with SF: {sf}, CR: {cr}, Tx Power: {tx_power}, ToA: {toa}, Packet size: {packet_size}")
        
        error_map = [1 if data[i] != reference_data[i] else 0 for i in range(min(len(data), len(reference_data)))]
        print(error_map)

        total_bytes = len(error_map)
        errors = sum(error_map)
        print(f"Total number of differences: {errors}")
        print(f"Bit error rate: {errors / total_bytes* 100:.2f}%")

        # Plot the error heatmap
        plot_error_heatmap_with_packets(error_map, packet_size, test_name=f"SF: {sf}, CR: {cr}, Tx: {tx_power}")

file_path = 'Payload_Data/100m_no_occlusion.txt'

compare_and_plot_experiments_with_dynamic_packet_size(file_path, data_og)
```
